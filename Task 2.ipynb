**Dropping irrelevant columns**

```python
# Deleting the column
col = ['Postal Code']
sample1 = sample.drop(columns=col,axis=1)
```

**Checking statistical relation between the various rows & columns**

```python
# Correlation between variables
sample1.corr()
```

```python
# Covariance of columns 
sample1.cov()
```

```python
sample1.head() # loads the first five rows
```

**Data Visualization**

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=sample1)
plt.title('Category vs Sub Category')
plt.xlabel('Sub-Catgory')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
sample1.corr() # Checking the correlation
```

```python
sample1.hist(bins=50,figsize=(20,15))
plt.show();
```

```python
# Count the total repeatable states
sample1['State'].value_counts()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=sample1['State'])
plt.xticks(rotation=90)
plt.title("STATE")
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=sample1['Category'])
plt.xticks(rotation=90)
plt.title("Category")
plt.show()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=sample1['Sub-Category'])
plt.xticks(rotation=90)
plt.title("Sub-Category")
plt.show()
```

```python
sls_prf=sample.groupby('Sub-Category')['Sales','Profit'].agg(['sum'])
sls_prf
```

```python
plt.style.use('bmh')
plt.figure()
sls_prf.plot(kind='bar',figsize=(20,8))
```
